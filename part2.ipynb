```python

import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(32, 8), subplot_kw=dict(aspect="equal"))

data = df['count(text)']
continent = df['continent']


def func(pct, allvals):
    absolute = int(pct/100.*np.sum(allvals))
    return "{:.1f}%\n({:d} )".format(pct, absolute)


wedges, texts, autotexts = ax.pie(data, autopct=lambda pct: func(pct, data),
                                  textprops=dict(color="w"),colors=["blue","coral","green","gray","purple","brown","pink","darkgray"])

ax.legend(wedges, continent,
          title="Continent",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

plt.setp(autotexts, size=10, weight="bold")

ax.set_title("Count tweet in each continent")

plt.show()



```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Load the world map shapefile
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Filter the world GeoDataFrame to select only the data for France
# france = world[world['name'] == 'France']

countries = ['Hong Kong', 'Singapore', 'United Arab Emirates', 'Philippines', 'India',
    'New Zealand', 'Australia', 
    'Canada', 'United States of America', 'Mexico',
    'Ghana', 'Uganda', 'Kenya', 'South Africa', 'France', 'Germany',
    'Ireland', 'Switzerland', 'United Kingdom']

colors = ['pink', 'pink','pink', 'pink', 'pink', 
          'brown', 'brown',
          'green', 'green', 'green',
          'coral', 'coral','coral','coral',
          'blue','blue','blue','blue','blue']

# Create a dictionary to store GeoDataFrames for each country
country_geodataframes = {}

# Filter and create GeoDataFrames for each country



# Create subplots
fig, ax = plt.subplots(1, 1, figsize=(12, 8))

# Plot the world map
world.boundary.plot(ax=ax, linewidth=0.8)

# Fill the interior of France with a color
# france.plot(ax=ax, color='red', edgecolor='black')
for country_name in countries:
    country = world[world['name'] == country_name]
    country.plot(ax=ax, color=colors.pop(0), edgecolor='black')

# Set the title

# Turn off the axis
ax.set_axis_off()

# Display the map
plt.show()
```
